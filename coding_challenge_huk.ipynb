```python
import sklearn
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
df['AreaEncoded'] = encoder.fit_transform(df['Area']) 
```

```python
df.head()
```

```python
df['VehBrandEncoded'] = encoder.fit_transform(df['VehBrand']) 
df['VehGasEncoded'] = encoder.fit_transform(df['VehGas']) 
```

```python
df.head()
```

```python
df['RegionEncoded'] = encoder.fit_transform(df['Region']) 
```

```python
df.head()
```

```python
df.info()
```

```python
X = df.drop(['Area', 'VehBrand', 'VehGas', 'Region', 'CAdivExp', 'IDpol', 'ClaimNb', 'Exposure'],axis=1)
```

```python
X.info()
```

```python
y = df["CAdivExp"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=1)
```

```python
from sklearn import linear_model

model = linear_model.TweedieRegressor(power=1.5, solver='newton-cholesky',max_iter=1000).fit(X_train,y_train)
model.score(X_train,y_train)
```

```python
#predict
y_pred = model.predict(X_test)
```

```python
#Evaluation 
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
print("Mean Square Error", mean_squared_error(y_pred,y_test))
print("Mean Absolute Error", mean_absolute_error(y_pred,y_test))
```

```python
from sklearn.metrics import d2_tweedie_score

d2_tweedie_score(y_test, y_pred, power=1.5)
```
